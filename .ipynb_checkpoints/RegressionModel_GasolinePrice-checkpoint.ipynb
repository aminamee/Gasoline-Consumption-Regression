## Multiple Regression Model - Full Model of log(mpg)

```python
list(log_df.columns.values)
```

```python
X = log_df[['displacement', 'horsepower', 'torque', 'compression_ratio', 'rear_axle_ratio', 'carburetor_barrels', 'transmission_speeds', 'overall_length', 'width', 'weight', 'transmission_type']]
y = log_df['mpg']

## fit a OLS model with intercept
X = sm.add_constant(X)
lm1 = smf.OLS(y, X).fit()

lm1.summary()
```

## Multiple Linear Regression - Subset Model of log(mpg)

```python
def forward_selected(data, response):
    """Linear model designed by forward selection.

    Parameters:
    -----------
    data : pandas DataFrame with all possible predictors and response

    response: string, name of response column in data

    Returns:
    --------
    model: an "optimal" fitted statsmodels linear model
           with an intercept
           selected by forward selection
           evaluated by adjusted R-squared
    """


    remaining = set(data.columns)
    remaining.remove(response)
    selected = []
    current_score, best_new_score = 0.0, 0.0
    while remaining and current_score == best_new_score:
        scores_with_candidates = []
        for candidate in remaining:
            formula = "{} ~ {} + 1".format(response,
                                           ' + '.join(selected + [candidate]))
            score = smf.ols(formula, data).fit().rsquared_adj
            scores_with_candidates.append((score, candidate))
        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates.pop()
        if current_score < best_new_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            current_score = best_new_score
    formula = "{} ~ {} + 1".format(response,
                                   ' + '.join(selected))
    model = smf.ols(formula, data).fit()
    return model

# source: http://trevor-smith.github.io/stepwise-post/
```

```python
lm2 = forward_selected(log_df, 'mpg')
lm2.summary()
```

```python
#qqnorm plot
fig = sm.qqplot(lm2.resid, fit=True, line='45')
plt.show
plt.title('QQ Norm Plot of Residuals')
```

```python
#fitted values vs residuals
plt.scatter(lm2.fittedvalues, lm2.resid)
plt.xlabel('Predicted MPG')
plt.ylabel('Residuals')
plt.title('Predicted MPG vs. Residuals')
```

## Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
log_df.head()
```

```python
pca_variables_df = gasoline[['displacement', 'horsepower', 'torque', 'compression_ratio', 'rear_axle_ratio', 'carburetor_barrels', 'transmission_speeds', 'overall_length', 'width', 'weight', 'transmission_type']]

# standard scores have zero mean and unit standard deviation
from sklearn import preprocessing
pca_data = preprocessing.scale(pca_variables_df.as_matrix())
```

```python
pca = PCA()
```

```python
pca.fit(pca_data)
print pca.explained_variance_
print pca.explained_variance_ratio_
print pca.explained_variance_ratio_.cumsum()
```

```python
pca_explained_variance = pca.explained_variance_ratio_

# show the scree plot for the pricipal component analysis
plt.bar(np.arange(len(pca_explained_variance)), pca_explained_variance, 
    color = 'grey', alpha = 0.5, align = 'center')
plt.title('PCA Proportion of Total Variance')  
```

```python
# compute full set of principal components (scores)
C = pca.transform(pca_variables_df)
```

```python
print pca.n_components_
```

```python
# add principal component scores to the original data frame
log_df['c1'] = C[:,0]
log_df['c2'] = C[:,1]
log_df['c3'] = C[:,2]
log_df['c4'] = C[:,3]
log_df['c5'] = C[:,4]
log_df['c6'] = C[:,5]
log_df['c7'] = C[:,6]
log_df['c8'] = C[:,7]
log_df['c9'] = C[:,8]
log_df['c10'] = C[:,9]
log_df['c11'] = C[:,10]
```

```python
# show variable loadings on principal components
loadings = pca.components_
print(loadings)
```

```python
# explore relationships between pairs of variables in the model
# that we are using for principal components regression
pca_model_data = log_df.loc[:,['mpg','c1','c2', 'c3', 'c4']]
```

```python
pca_model_data.corr().round(decimals=3)
```

```python
import seaborn as sb
# show scatter plot matrix for the data used in the linear model
#scatter_matrix(pca_model_data, alpha=0.2, figsize=(6, 6), diagonal='kde')
sb.pairplot(pca_model_data)
```

## Regress a Linear Model on the first PCA Component (~70% of variation)

```python
# fit regression model using original variables

# define pca model for regression using first component
pca_model = str('mpg ~ c1')


# that is, regress the response y on the first two principal components
pca_model_fit = smf.ols(pca_model, data = pca_model_data).fit()
# summary of model fit to the training set
print(pca_model_fit.summary())
```

```python
# training set predictions from the model fit to the training set
log_df['pca_y'] = pca_model_fit.fittedvalues

# compute the proportion of response variance accounted for by pca model
# this should match the value obtained from the model summary
print('\nProportion of Variance Accounted for by Principal Components Regression: ',
    np.power(log_df['mpg'].corr(log_df['pca_y']),2).round(decimals = 3))

# Note. As far as the proportion of variance acconted for,     
# the regression on the first two pricipal components does almost as well
# the regression on the original five explanatory variables.
```

```python
#fitted values vs residuals
plt.scatter(pca_model_fit.fittedvalues, pca_model_fit.resid)
plt.xlabel('Predicted MPG')
plt.ylabel('Residuals')
plt.title('PCA Predicted MPG vs. Residuals')
```

```python
#qqnorm plot
fig = sm.qqplot(pca_model_fit.resid, fit=True, line='45')
plt.show
plt.title('QQ Norm Plot of Residuals')
```

## Regress a linear model on the first four Principal Components (~95% of variation)

```python
# fit regression model using original variables

# define pca model for regression using first component
pca_model2 = str('mpg ~ c1 + c2 + c3 + c4')


# that is, regress the response y on the first two principal components
pca_model_fit2 = smf.ols(pca_model2, data = pca_model_data).fit()
# summary of model fit to the training set
print(pca_model_fit2.summary())
```

```python
# training set predictions from the model fit to the training set
log_df['pca_y2'] = pca_model_fit2.fittedvalues

# compute the proportion of response variance accounted for by pca model
# this should match the value obtained from the model summary
print('\nProportion of Variance Accounted for by Principal Components Regression: ',
    np.power(log_df['mpg'].corr(log_df['pca_y2']),2).round(decimals = 3))

# Note. As far as the proportion of variance acconted for,     
# the regression on the first two pricipal components does almost as well
# the regression on the original five explanatory variables.
```

```python
#fitted values vs residuals
plt.scatter(pca_model_fit2.fittedvalues, pca_model_fit2.resid)
plt.xlabel('Predicted MPG')
plt.ylabel('Residuals')
plt.title('PCA Predicted MPG vs. Residuals')
```

```python
#qqnorm plot
fig = sm.qqplot(pca_model_fit.resid, fit=True, line='45')
plt.show
plt.title('QQ Norm Plot of Residuals')
```
